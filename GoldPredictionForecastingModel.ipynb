## ` Functions to normalize the outliers in the columns.`

````python
# Function to normalize outliers

def outlier_removal(column):
    # Capping the outlier rows with Percentiles
    upper_limit = column.quantile(.95)

    # set upper limit to 95percentile
    lower_limit = column.quantile(.05)

    # set lower limit to 5 percentile
    column.loc[(column > upper_limit)] = upper_limit
    column.loc[(column < lower_limit)] = lower_limit
    return column


# Normalize outliers in columns except Date

data_new[['SPX', 'GLD', 'USO', 'SLV', 'EUR/USD']] = \
    data_new[['SPX', 'GLD', 'USO', 'SLV',  'EUR/USD']].apply(outlier_removal)
````

# **`# LSTM (Long Short-Term Memory) Model`**

It is particularly well-suited for capturing long-range dependencies in sequential data, making it an ideal choice for tasks such as time series forecasting. In this context, LSTM models can be employed to forecast gold prices, leveraging machine learning techniques to analyze historical trends and make predictions about future price movements.

````python
model = Sequential([
    LSTM(50, return_sequences= True, input_shape= (x_train.shape[1], n_cols)),
    LSTM(64, return_sequences= False),
    Dense(32),
    Dense(16),
    Dense(n_cols)
])

model.compile(optimizer= 'adam', loss= 'mse' , metrics= "mean_absolute_error")
````

````python
# Fitting the LSTM to the Training set
history = model.fit(x_train, y_train, epochs= 100, batch_size= 32)
````
